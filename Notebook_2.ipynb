The missing values in the % change in FTAs is trivial and can be calculated as follows:

```python
for i in range(1, len(df_fta_combined)):
    if pd.isna(df_fta_combined.loc[i, 'Percentage (%) change over previous year']):
        df_fta_combined.loc[i, 'Percentage (%) change over previous year'] = ((df_fta_combined.loc[i, 'FTAs in India'] - df_fta_combined.loc[i - 1, 'FTAs in India']) * 100) / df_fta_combined.loc[i - 1, 'FTAs in India']
df_fta_combined.head(23)
```

The missing values in the Fee % change over previous year can be calculated as follows:

```python
for i in range(1, len(df_fta_combined)):
    if pd.isna(df_fta_combined.loc[i, 'Fee % Change over previous year']):
        df_fta_combined.loc[i, 'Fee % Change over previous year'] = ((df_fta_combined.loc[i, 'FEE through tourism (in Crores)'] - df_fta_combined.loc[i - 1, 'FEE through tourism (in Crores)']) * 100) / df_fta_combined.loc[i - 1, 'FEE through tourism (in Crores)']

df_fta_combined['Fee % Change over previous year'] = df_fta_combined['Fee % Change over previous year'].round(1)
df_fta_combined.head(23)
```

```python
df_copy = df_fta_combined.copy()

df_copy.isnull().sum()
```

```python
# Columns representing age groups
age_group_columns = ['0-14', '15-24', '25-34', '35-44', '45-54', '55-64', '65 and above']
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare data for regression
historical_data = df_copy[df_copy['Year'] <= 2019]
years = historical_data['Year'].values.reshape(-1, 1)

# Predict missing values for each column
for col in age_group_columns:
    model = LinearRegression()
    model.fit(years, historical_data[col].values)

    # Predict for 2022 and 2023
    predicted_values = model.predict(np.array([2020, 2021, 2022, 2023]).reshape(-1, 1))
    df_copy.loc[df_copy['Year'].isin([2020, 2021, 2022, 2023]), col] = predicted_values

# Normalize the rows to ensure they sum to 100%
df_copy[age_group_columns] = df_copy[age_group_columns].apply(
    lambda row: row / row.sum() * 100 if row.sum() != 100 else row,
    axis=1
)

df_copy[age_group_columns] = df_copy[age_group_columns].round(1)
df_copy
```

```python
df_copy.isnull().sum()
```

```python
# List of age group columns
quarter_columns = ['Q1 (Jan-Mar)', 'Q2 (Apr-June)', 'Q3 (July-Sep)', 'Q4 (Oct-Dec)']

# Calculate mean proportions for prior years (e.g., up to 2019)
mean_proportions = df_copy.loc[df_copy['Year'] <= 2019, quarter_columns].mean()

# Impute missing values for 2020-2023
def fill_with_proportions(row):
    if row[quarter_columns].isnull().any():
        row[quarter_columns] = mean_proportions
    return row

df_copy = df_copy.apply(fill_with_proportions, axis=1)
df_copy[quarter_columns] = df_copy[quarter_columns].round(1)
df_copy
```

```python
df_copy.isnull().sum()
```
